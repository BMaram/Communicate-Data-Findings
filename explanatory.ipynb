## Dataset Overview

> The dataset consists of 3854 movies with 16 variables on each movie, collected from The Movie Database (TMDb), including popularity and revenue and budget.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
baseColor = sb.color_palette()[0]
# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv("tmdb-movies_clean.csv")
```

## What is the most common movie genres?

> Here is a bar chart illustrating the number of movies in each genre.

```python
plt.figure(figsize = [20, 10])
type_counts = df['genre'].value_counts()
type_order = type_counts.index
sb.countplot(data = df , y='genre' , color = baseColor, order = type_order);
```

 > Overall we can see that Drama then Comedy, Thriller and Action are the most commen generes in this dataset.

## How popularity affects revenue?

>Here is a scatter plot with fitting line shows the relation between popularty and revenue.

```python
plt.figure(figsize = [20, 10])
sb.regplot(x="popularity", y="revenue",data=df);
```

> We can see there is a positive relationship between popularity and revenue

## How has the popularity of most popular generes changed over years?

> here is a colored line plot to show relation between popularity and release_year for each movie genere.


```python
plt.figure(figsize = [20, 10])
genere_df = df.loc[(df.genre=='Drama')|(df.genre=='Comedy')]
sb.lineplot(x="release_year", y="popularity", hue="genre", estimator=None,data=genere_df);
```

> we can see that the popularity on both drama and comedy incresed over years but more on Drama.
